Ejercicio 2: Desarrolle una gráfica de linea usando el módulo pyplot de Matplotlib para producir la función:

$$ f(x) = x^2 |sen(x)| - 4 $$

Sobre el dominio $ 0 \leq x \leq 4 $. Asegurese de que los ejes y la curva esté correctamente etiquetadas. Agregue un titulo. 

```python
import numpy as np

# Define the function
def f(x):
    return x**2 * np.abs(np.sin(x)) - 4

# Define the domain
x = np.linspace(0, 4, 400)
y = f(x)
```

```python
# Create the plot
plt.figure()
plt.plot(x, y, label='$f(x) = x^2 |sen(x)| - 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de $f(x) = x^2 |sen(x)| - 4$')
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
# Definir el dominio
x = np.linspace(-1, 1, 400)

# Calcular las funciones hiperbólicas
sinh_x = np.sinh(x)
cosh_x = np.cosh(x)
tanh_x = np.tanh(x)
sech_x = 1 / np.cosh(x)

# Crear la figura y los ejes
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Gráfico de senh(x) vs x
axs[0, 0].plot(x, sinh_x, label='$senh(x)$')
axs[0, 0].set_title('$senh(x) \; vs \; x$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('$senh(x)$')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Gráfico de cosh(x) vs x
axs[0, 1].plot(x, cosh_x, label='$cosh(x)$', color='r')
axs[0, 1].set_title('$cosh(x) \; vs \; x$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('$cosh(x)$')
axs[0, 1].grid(True)
axs[0, 1].legend()

# Gráfico de tanh(x) vs x
axs[1, 0].plot(x, tanh_x, label='$tanh(x)$', color='g')
axs[1, 0].set_title('$tanh(x) \; vs \; x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('$tanh(x)$')
axs[1, 0].grid(True)
axs[1, 0].legend()

# Gráfico de sech(x) vs x
axs[1, 1].plot(x, sech_x, label='$sech(x)$', color='m')
axs[1, 1].set_title('$sech(x) \; vs \; x$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('$sech(x)$')
axs[1, 1].grid(True)
axs[1, 1].legend()

# Ajustar el layout
plt.tight_layout()
plt.show()
```
